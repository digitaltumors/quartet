```python
quartet_maf = pd.read_csv('/cellar/projects/ccle_qc/quartet_output/Quartet_v2.0/Quartet_v2.0.strelka2.prepopfilt.maf', sep = '\t')
```

```python
gnomade_pass = np.logical_or(quartet_maf['gnomADe_AF'] < 1e-5, quartet_maf['gnomADe_AF'].isna())
quartet_maf = quartet_maf.loc[gnomade_pass]
```

```python
quartet_maf = quartet_maf.loc[quartet_maf['Chromosome'].apply(lambda x: x in chromosome_mapping)]
quartet_maf['Chromosome'] = quartet_maf['Chromosome'].apply(lambda x: chromosome_mapping[x])
quartet_maf.to_csv('/cellar/projects/ccle_qc/quartet_output/Quartet_v2.0/Quartet_v2.0.strelka2.maf', sep = '\t', index = False)
```
